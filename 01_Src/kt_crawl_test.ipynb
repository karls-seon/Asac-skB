```python
from bs4 import BeautifulSoup

first_table = soup.select_one('table.plan-list')

if first_table:
    plan_name = first_table.select_one('th.title strong').text.strip()
    price = first_table.select_one('td.charge strong').text.strip()
    features_tags = first_table.select('.plan-info ul.plans li p')
    features = [tag.text.strip().replace('\n', ' ') for tag in features_tags]

    print(f"📌 요금제명: {plan_name}")
    print(f"💰 요금: 월 {price}원")
    print(f"🎁 혜택: {', '.join(features)}")

else:
    print("테이블을 찾을 수 없습니다.")
```

```python
item_codes = []

tags_with_itemcode = soup.select('a[href*="ItemCode="]')

for tag in tags_with_itemcode:
    href = tag['href']
    
    # 정규표현식을 사용하여 ItemCode= 뒤에 있는 숫자만 추출
    match = re.search(r'ItemCode=(\d+)', href)
    
    if match:
        item_code = match.group(1)
        
        # 중복 방지를 위해 리스트에 없는 경우만 추가 (선택 사항)
        if item_code not in item_codes:
            item_codes.append(item_code)

print("추출된 ItemCode 목록:", item_codes)
```

```python
import requests
import re

item_code = "1691"
base_url = "https://product.kt.com"
item_url = f"https://product.kt.com/wDic/productDetail.do?ItemCode={item_code}&CateCode=6002&filter_code=186&option_code=255&pageSize=10"

# 1. 메인 페이지 HTML 가져오기
response = requests.get(item_url)
html_text = response.text

# 2. 정규식 수정: 파일명(htmlUploadType_숫자.html) 부분만 정확히 쏙 빼옵니다.
match = re.search(r'(htmlUploadType_\d+\.html)', html_text)

if match:
    file_name = match.group(1) # 찾은 파일명: htmlUploadType_20260721155852.html
    
    # 3. 우리가 알고 있는 ItemCode와 조합해서 완벽한 URL을 만듭니다.
    final_detail_url = f"{base_url}/static/prodetail/{item_code}/web/{file_name}"
    
    print(f"🎉 조립 완료된 세부페이지 주소:\n{final_detail_url}")
    
    # 4. 이제 진짜 데이터가 있는 세부페이지로 접속!
    detail_html = requests.get(final_detail_url).text
    # print(detail_html) # 여기서 BeautifulSoup으로 크롤링하시면 됩니다.
    
else:
    print("파일명을 찾을 수 없습니다.")
```
