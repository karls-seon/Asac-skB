```python
try:
    with open("epost_api_key.txt", "r") as file:
        api_key = file.read().strip() 
except FileNotFoundError:
    print("epost_api_key.txt 확인!!")
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

BASE_URL = "http://openapi.epost.go.kr/postal/retrieveAlddlPhoneService/retrieveAlddlPhoneService/getAlddlPhoneList"

params = {
    "ServiceKey": api_key,  # 필수: 공공데이터포털에서 발급받은 인증키
}

res = requests.get(BASE_URL, params=params)
soup = BeautifulSoup(res.text, "xml")
```

```python
items = []
for item in soup.find_all("alddlPhone"):
    items.append({
        "telecom_name": item.find("telecomName").text,
        "biz_name": item.find("bizName").text,
        "dispyn": item.find("dispyn").text,
        "phone_name": item.find("phoneName").text,
        "model_name": item.find("modelName").text,
        "phone_generation_type": item.find("phoneGenerationType").text,
        "phone_shape_type": item.find("phoneShapeType").text,
        "brand_name_type": item.find("brandNameType").text,
        "phone_used_div": item.find("phoneUsedDiv").text if item.find("phoneUsedDiv") else None,
        "phone_total_price": item.find("phoneTotalPrice").text,
        "phone_div_month": item.find("phoneDivMonth").text,
        "div12prc": item.find("div12prc").text if item.find("div12prc") else None,
        "div24prc": item.find("div24prc").text if item.find("div24prc") else None,
        "div30prc": item.find("div30prc").text if item.find("div30prc") else None,
        "change_phone_yn": item.find("changePhoneYn").text if item.find("changePhoneYn") else None,
    })

df = pd.DataFrame(items)
```

```python
column_labels = {
    "telecom_name": "통신망",
    "biz_name": "업체명",
    "dispyn": "게시여부",
    "phone_name": "단말기명",
    "model_name": "모델명",
    "phone_generation_type": "통신망타입",
    "phone_shape_type": "단말기형태",
    "brand_name_type": "제조사",
    "phone_used_div": "신품구분",
    "phone_total_price": "할부원가(원)",
    "phone_div_month": "할부개월수",
    "div12prc": "12개월할부금액",
    "div24prc": "24개월할부금액",
    "div30prc": "30개월할부금액",
    "change_phone_yn": "기기변경여부",
}

df_kr = df.rename(columns=column_labels)

print(f"=== 알뜰폰 단말기 조회 결과 ({len(df_kr)}건) ===")
df_kr
```
