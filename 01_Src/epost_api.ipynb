```python
try:
    with open("epost_api_key.txt", "r") as file:
        api_key = file.read().strip() 
except FileNotFoundError:
    print("epost_api_key.txt 확인!!")
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

BASE_URL = "http://openapi.epost.go.kr/postal/retrieveAlddlChargeService/retrieveAlddlChargeService/getAlddlChargeList"

params = {
    "ServiceKey": api_key,  # 필수: 공공데이터포털에서 발급받은 인증키
}

res = requests.get(BASE_URL, params=params)
soup = BeautifulSoup(res.text, "xml")
```

```python
column_labels = {
    "telecomName": "통신망",
    "bizName": "업체명",
    "chargeName": "요금제명",
    "telecomGenerationType": "통신상품구분",
    "chargeDiv": "요금제구분",
    "chargeAmount": "기본료(원)",
    "voiceAmount": "기본음성(분)",
    "messageAmount": "기본문자(건)",
    "dataAmount": "기본데이터(MB)",
    "overVoiceAmtUnit": "초과음성단가(원/초)",
    "overMessageAmtUnit": "초과문자단가(원/건)",
    "overDataAmtUnit": "초과데이터단가(원/MB)",
    "div12prc": "12개월할부금액",
    "div24prc": "24개월할부금액",
}

items = []
for charge in soup.find_all("alddlCharge"):
    row = {}
    for eng_key, kor_key in column_labels.items():
        tag = charge.find(eng_key)
        row[kor_key] = tag.text.strip() if tag and tag.text else None
    items.append(row)

df = pd.DataFrame(items)
df
```
